```python
#import matplotlig lib
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

Ploting the data

```python
# sin wave
plt.plot(x,y)
plt.show()
```

```python
# cosine wave
plt.plot(x,z)
plt.show()
```

adding tittle X-axis and Y-axis on the Graph

```python
plt.plot(x,y)
plt.xlabel('X-angle')
plt.ylabel('Y-sine wave')
plt.title("sine wave")
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) # 0,0 represent origin and 1,1 represent hight width

languages = ['English','French', 'Spanish', 'Latin', 'Gernman']
people = [100,50,10,40,70]
ax.bar(languages,people)
plt.xlabel("LANGUAGES")
plt.ylabel('NUMBER OFPEOPLE')
plt.show()
```

**Pie Chart**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French', 'Spanish', 'Latin', 'Gernman']
people = [100,50,10,40,70]
ax.pie(people , labels=languages, autopct='%1.1f%%')
plt.show()
```

**Scatter Plot**

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z= 20*np.random.random(100)
x= np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()

```
